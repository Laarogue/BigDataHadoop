```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
data = pd.read_csv("data/house_pricing.csv",sep=",")
data.head()
```

```python
data.info()
```

```python
sns.relplot(x='LotArea',y='SalePrice',data=data)
```

```python
data2 = data[(data.LotArea < 20000) & (data.SalePrice < 500000)]
sns.relplot(x='LotArea',y='SalePrice',data=data2)
```

```python
sns.relplot(x='LotFrontage',y='LotArea',data=data)
```

```python
data2 = data[(data.LotArea < 100000) & (data.LotFrontage < 200)]
sns.relplot(x='LotFrontage',y='LotArea',data=data2)
```
